```python
import numpy as np
import pandas as pd
#data = pd.read_csv('C:/Users/varunn/Documents/Confluence attachments/MachinelearningHeadcount.csv')
data1 = pd.read_csv('C:/Users/varunn/Documents/Confluence attachments/Placeholder job analysis/Machinelearningv2.csv')
features_raw = data1.drop(['jobid'],axis = 1)
features = pd.get_dummies(features_raw,columns=['ClientID'])
```

```python
import matplotlib.pyplot as plt

# Create data

x = (data1['ClientID'])
y = data1['Headcount']
colors = (0,0,0)
area = np.pi*3
 
# Plot
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('Scatter plot revenue vs headcount')
plt.xlabel('FinanceValue')
plt.ylabel('Client')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D



fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

xs = (data1['ClientID'])
ys = data1['Headcount']
zs = np.log(data1['FinanceValue'])*0

color = ("red", "green", "blue")

xt = (data1['ClientID'])
yt = (data1['Headcount'])*0
zt = (np.log(data1['FinanceValue'])) 

ax.scatter(xs, ys, zs, c='g', marker='o')

ax.scatter(xt, yt, zt, c='b', marker='^')

ax.set_xlabel('Client')
ax.set_ylabel('Headcount')
ax.set_zlabel('Finance')

plt.show()
```

```python
from matplotlib import pyplot as plt
import matplotlib.mlab as mlab
array = features['Headcount']

n, bins, patches = plt.hist(array,50, normed=1)
mu = np.mean(array)
sigma = np.std(array)
plt.plot(bins, mlab.normpdf(bins, mu, sigma))
plt.show()
```
